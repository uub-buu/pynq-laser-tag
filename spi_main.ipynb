```python
%%microblaze base.PMODB
#include <stdint.h>
char tx = 0x00;
char rx = 0x00;

#include "spi.h"
int spi_init(){
    spi spi_dev = spi_open(0, 1, 2, 3);
    spi_transfer(spi_dev, &tx, &rx, 1);
    return (int)rx;
}
```

```python
#loop through and get data being sent
while True:
    x = spi_init()
    print(f"TX: {x}")
```
